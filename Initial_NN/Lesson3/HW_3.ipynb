```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
pd.set_option('display.max_rows', None)
```

```python
data = load_boston()
```

### Подготовим данные

```python
scaler = StandardScaler()
X = scaler.fit_transform(data['data'])
```

```python
X = pd.DataFrame(X, columns=data['feature_names'])
y =  pd.DataFrame(data=data['target'], columns=['target'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

### Создадим автоматический gridsearch на все параметры

```python
def grid_test_models_1(train_data,  test_data, train_target, test_target, optimizer='adam' ,base_neurons=128, add_layers=0, activation='linear', metric='MSE', loss='MSE' ):
    model = Sequential()
    model.add(Dense(base_neurons, activation=activation, input_shape=(train_data.shape[1],)))  
    for k in range(add_layers):
        model.add(Dense(10+(10-k)*10, activation=activation))       
    model.add(Dense(1, activation='linear')) 

    model.compile(
      optimizer=optimizer,
      loss=loss,
      metrics=metric,
    )  
    
    model.fit(    
      train_data,
      train_target,
      epochs=200,
      batch_size=100, validation_split=0.2, verbose=0
    )
    
    eval_metric = model.evaluate(test_data, test_target, verbose=0)[1]
    preds = model.predict(test_data)
    r2_sc = r2_score(y_test, preds)
    result = {'optimizer' : optimizer, 'base_neurons' : base_neurons, 'layers' : 2 + add_layers, 'activation' : activation, 'loss' : loss, 'metric' : metric , 'eval_metric' : eval_metric, 'Метрика R2' :  r2_sc}
    return result
      
```

- Создадим таблицу для записи результатов

```python
table = {'optimizer' : [], 'base_neurons' : [], 'layers' : [], 'activation' : [], 'loss' : [], 'metric' : [], 'eval_metric' : [], 'Метрика R2' : []}
```

- Создадим сетку с параметрами

```python
grid_param = {'optimizer' : ['adam', 'AdaGrad', 'SGD', 'RMSprop'],
              'base_neurons' : [20,50,80,128,300],
              'add_layers' : [0,1,2,3,4,5],
              'activation' : ['linear', 'elu', 'relu', 'tanh', 'sigmoid'],
              'loss' : ['MAE', 'MSE', 'Huber', 'LogCosh'],
              'metric' : ['MeanAbsolutePercentageError', 'MAE', 'MSE']}
```

- Запустим поиск по сетке

```python
for opt in grid_param['optimizer']:
    for neur in grid_param['base_neurons']:
        for ls in grid_param['add_layers']:
            for act in grid_param['activation']:
                for loss in grid_param['loss']:
                    for met in grid_param['metric']:        
                        spam = grid_test_models_1(X_train, X_test, y_train, y_test, opt, neur, ls, act, met, loss)
                        for i in table:
                            table[i].append(spam[i])
```

```python
pd.DataFrame(table).head(10)
```

- Сохраним результаты

```python
results = pd.DataFrame(table)
```

```python
results.to_csv('results.csv')
```

### Посмотрим, как училась модель с лучшими показателями

```python
model = Sequential()
model.add(Dense(300, activation='relu', input_shape=(X_train.shape[1],)))  
for k in range(6):
    model.add(Dense(10+(10-k)*10, activation='relu'))       
model.add(Dense(1, activation='linear')) 

model.compile(
  optimizer='adam',
  loss='Huber',
  metrics='MSE',
)  

history = model.fit(    
  X_train,
  y_train,
  epochs=200,
  batch_size=100, validation_split=0.2
)
```

```python
preds = model.predict(X_test)
```

```python
r2_score(y_test, preds)
```

```python
plt.plot(history.history['val_MSE'] ,label='val')
plt.plot(history.history['MSE'] ,label='train')
plt.legend()
plt.xlabel('epoch')
plt.ylabel('MSE')
plt.title('точность сети от эпох')
plt.grid()
plt.show()
```
